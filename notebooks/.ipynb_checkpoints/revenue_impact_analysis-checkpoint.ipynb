# Problem Statement
## Pens and Printers
Pens and Printers is a company that provides high quality office products to large organizations. Six weeks ago Pens and Printers launched a new line of office stationery. Despite the world becoming increasingly digital, there is still demand for notebooks, pens and sticky notes.

Launching a new product line is expensive and we need to make sure we are using the best techniques to sell the new product effectively. The best approach may vary for each new product so we need to learn quickly what works and what doesn’t.

Our focus has been on selling products to enable our customers to be more creative, focused on tools for brainstorming. We have tested three different sales strategies for this, targeted email and phone calls, as well as combining the two.
- *Email:* Customers in this group received an email when the product line was launched, and a further email three weeks later. This required very little work for the team.
- *Call:* Customers in this group were called by a member of the sales team. On average members of the team were on the phone for around thirty minutes per customer.
- *Email and call:* Customers in this group were first sent the product information email, then called a week later by the sales team to talk about their needs and how this new product may support their work. The email required little work from the team, the call was around ten minutes per customer.

## Exploratory Data Analysis
In this section, we explore the dataset to understand its structure, 
identify missing values, and detect potential anomalies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('../data/product_sales.csv')
df.head(3)
```

### At first glance, we can see that the data types are as expected.

```python
df.dtypes
```

### We verified that the "week" column does not contain null values. We also observed that the proportions are higher in the earlier weeks.

```python
print('NaN counter: ', df['week'].isna().sum())
print('Weekly sales proportion: \n', df['week'].value_counts(normalize=True))
```

### We verified that the "sales_method" column does not contain null values. We need to standardize the column values because there appear to be some typographical errors.

```python
print('NaN counter: ', df['sales_method'].isna().sum())
print('Sales method proportion: \n', df['sales_method'].value_counts(normalize=True))
```

### We defined a function to standardize the "sales_method" values.

```python
def clean_sales_method(method):
    if method == 'em + call':
        return 'Email + Call'
    elif method == 'email':
        return 'Email'
    return method

df['sales_method_clean'] = df['sales_method'].apply(clean_sales_method)
```

### Take a look to verify that the function worked as expected.

```python
print(df['sales_method'].value_counts(), '\n')
print(df['sales_method_clean'].value_counts())
```

### "customer_id" is supposed to be an identifier, so it should not contain NA values or blank entries.

```python
print(df['customer_id'].isna().sum())
print(df[df['customer_id']=='']['customer_id'])
```

### The "nb_sold" column should not contain negative or NA values. We reviewed the basic descriptive statistics.

```python
print('NaN counter:', df['nb_sold'].isna().sum(), '\n')
print(df['nb_sold'].describe())
```

### The "revenue" column contains 1,074 missing values, which exceeds 5% of the data; therefore, we should not drop these records.

```python
print('NaN counter: ', df['revenue'].isna().sum())
```

### We analyzed the proportion of missing values for each of the three possible values in the "sales_method_clean" column.

```python
email_revenue_isna = df[df['sales_method_clean'] == 'Email']['revenue'].isna().sum() \
/ df[df['sales_method_clean'] == 'Email']['sales_method_clean'].count()

emailCall_revenue_isna = df[df['sales_method_clean'] == 'Email + Call']['revenue'].isna().sum() \
/ df[df['sales_method_clean'] == 'Email + Call']['sales_method_clean'].count()

call_revenue_isna = df[df['sales_method_clean'] == 'Call']['revenue'].isna().sum() \
/ df[df['sales_method_clean'] == 'Call']['sales_method_clean'].count()

print('Email % de NaN: ', email_revenue_isna)
print('Email+Call % de NaN: ', emailCall_revenue_isna)
print('Call % de NaN: ', call_revenue_isna)
```

### We performed median imputation by "sales_method_clean" category.

```python
method_sales_revenue = df.groupby('sales_method_clean')['revenue'].median()
revenue_dict = method_sales_revenue.to_dict()

df['revenue_clean'] = df['revenue'].fillna(df['sales_method_clean'].map(revenue_dict))

print('Contador NaN: ', df['revenue_clean'].isna().sum())
```

### The "years_as_customer" column does not contain negative values, but it includes unrealistically high values. Since the company was founded in 1984, the maximum possible value in 2026 is 42.

```python
print('NaN counter: ', df['years_as_customer'].isna().sum(), '\n')
print(df['years_as_customer'].describe())
```

### he "nb_site_visits" column does not contain NA values or negative values.

```python
print('NaN counter: ', df['nb_site_visits'].isna().sum(), '\n')
print(df['nb_site_visits'].describe())
```

### "state" column does not contain NA values or blank entries.

```python
print('NaN counter: ', df['state'].isna().sum())
print(df['state'].value_counts().sort_index())
```

## Answering the Business Questions

#### 1. How many customers were there for each approach?
The sales method with the highest client count is "Email", accounting for 7,466 clients, followed by "Call" with 4,962 clients, and "Email + Call" with 2,572 clients.

```python
sns.countplot(data=df, x='sales_method_clean')
plt.title('Count of customers by sales method')
plt.show()
```

#### 2. What does the spread of the revenue look like overall? And for each method?

```python
email = df[df['sales_method_clean']=='Email']['revenue_clean']
email_call = df[df['sales_method_clean']=='Email + Call']['revenue_clean']
call = df[df['sales_method_clean']=='Call']['revenue_clean']

fig, axes = plt.subplots(1, 4, figsize=(15, 4), sharey=True)

sns.histplot(df['revenue_clean'], ax=axes[0])
axes[0].set_title('Total revenue')

sns.histplot(email, ax=axes[1])
axes[1].set_title('Email')

sns.histplot(email_call, ax=axes[2])
axes[2].set_title('Email + Call')

sns.histplot(call, ax=axes[3])
axes[3].set_title('Call')


plt.suptitle('histograms by sales method')
plt.tight_layout()
plt.show()
```

#### based on the upper bound, revenue values exceeding 190 can be classified as outliers within the total revenue distribution.

```python
revenue_q75 = df['revenue_clean'].quantile(0.75)
revenue_q25 = df['revenue_clean'].quantile(0.25)
revenue_iqr = revenue_q75 - revenue_q25
print(f'Revenue IQR: {revenue_iqr}')
print(f'Lower bound: {revenue_q25 - 1.5*revenue_iqr}')
print(f'Upper bound: {revenue_q75 + 1.5*revenue_iqr}')
```

#### In the following boxplots, we can observe that the median revenue for 'Email + Call' is substantially higher than that of the other two methods. It is also important to note that although 'Email + Call' represents the smallest percentage of clients in the dataset, it has both the highest mean and median revenue.

```python
plt.clf()
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# KDE plot en el primer subplot
sns.boxplot(y=df['revenue_clean'], ax=axes[0])
axes[0].set_title("Total Revenue Distribution")

# Boxplot en el segundo subplot
sns.boxplot(data=df, x='sales_method_clean', y='revenue_clean', ax=axes[1])
axes[1].set_title("Revenue distribution by sales method")

# Ajustar espacios
plt.tight_layout()
plt.show()
```

#### 3. Was there any difference in revenue over time for each of the methods?
There is a noticeable difference in revenue over time. As shown, the "Email" method starts with high revenue figures but experiences a sharp decline as the weeks progress. In contrast, the "Email + Call" method shows a consistent increase in revenue over time, with a slight dip in week 6. Lastly, the "Call" method also shows a gradual increase in revenue over the weeks, although the week-to-week growth is minimal; it also exhibits a decline in week 6.

```python
plt.clf()
weekly_sales = df.groupby(['sales_method_clean','week'])['revenue_clean'].sum().reset_index()
sns.lineplot(data=weekly_sales, x='week', y='revenue_clean', hue='sales_method_clean')
plt.title('Revenue per week')
plt.show()
```

#### 4. Based on the data, which method would you recommend we continue to use? Some of these methods take more time from the team so they may not be the best for us to use if the results are similar.
I recommend continuing the use of the "Email" method, as it requires less effort from the team while also generating the highest revenue. The "Email + Call" method should also be maintained, since despite its relatively small share of clients in the dataset, it yields very high revenue. On the other hand, I recommend discontinuing the "Call" method and replacing it with "Email + Call". Considering that the average call duration for the "Call" method is around 30 minutes, whereas for "Email + Call" it is approximately 10 minutes, this would allow the team to handle up to three contacts via "Email + Call" for every one handled via "Call".
